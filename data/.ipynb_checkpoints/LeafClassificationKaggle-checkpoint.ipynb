```python
import sklearn.linear_model as lm
```

```python
lgr = lm.LogisticRegression(multi_class = 'multinomial', solver = 'newton-cg')
```

```python
y = dataset.species

from sklearn import preprocessing
le = preprocessing.LabelEncoder()
if y.dtype == object:
    y = le.fit_transform(y)
else:
    pass
```

```python
X = dataset.drop('species', axis=1)
X = (dataset.ix[:,1:].values).astype('float32')
```

```python
from sklearn.model_selection import train_test_split
```

```python
#X_train, X_test, y_train, y_test = train_test_split(X, y,test_size=0.2, random_state = 0)
from sklearn.model_selection import StratifiedShuffleSplit

sss = StratifiedShuffleSplit(n_splits = 1, test_size=0.2, random_state=0)
sss.get_n_splits(X, y)
X_train = []
X_test = []
for train_index, test_index in sss.split(X, y):    
    X_train, X_test, y_train, y_test = (X[train_index], X[test_index], y[train_index], y[test_index])
```

```python
lgr.fit(X_train, y_train)
```

```python
from sklearn.metrics import log_loss
```

```python
y_predicted = lgr.predict_proba(X_test)
if y_predicted.dtype == object:
    y_predicted = le.fit_transform(y_predicted)
else:
    pass
```

```python
logLoss = log_loss(y_test, y_predicted)
```

```python
print("Log Loss from multinomial logistics regression = " + str(logLoss))
```
